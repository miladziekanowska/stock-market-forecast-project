Made a dictionary that contains informations about Dow Jones stocks

```python
ticker_list = yh.tickers_dow()
historical_datas = {}
for ticker in ticker_list:
    historical_datas[ticker] = yh.get_data(ticker, start_date="13/07/1993", end_date="12/07/2023", index_as_date = False, interval="1d")
```

Percentage of change during day counted

```python
for key in historical_datas:
  historical_datas[key]['%changed'] = historical_datas[key].apply(lambda x: ((x[4] - x[1])/x[1] * 100), axis = 1)
```
